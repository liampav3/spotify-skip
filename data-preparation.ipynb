### Data Directory Structure
Should have a subdirectory named data that contains
1. One of the history csvs from the test set, named data.csv
2. A subdirectory named track_features containing both track feature csvs

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = pd.read_csv('data/data.csv')
data[:20]
```

```python
data.shape
```

```python
data = data.drop('date', axis=1)
```

```python
data['hist_user_behavior_reason_end'][data.hist_user_behavior_reason_end != 'trackdone'] = 1
data['hist_user_behavior_reason_end'][data.hist_user_behavior_reason_end == 'trackdone'] = 0
data[:10]
```

```python
data.replace({False: 0, True:1}, inplace=True)
data
```

```python
track_features1 = pd.read_csv('data/track_features/tf_000000000000.csv')
track_features2 = pd.read_csv('data/track_features/tf_000000000001.csv')
track_features = track_features1.append(track_features2)
```

```python
track_features.set_index('track_id', inplace=True)
track_features
```

```python
features_data = track_features.loc[data['track_id_clean']]
features_data = features_data.reset_index(drop=True)
features_data
```

```python
data = data.drop('track_id_clean', axis=1)
encoder = OrdinalEncoder()
columns = data.columns
data = encoder.fit_transform(data)
data = pd.DataFrame(data, columns=columns)
data
```

```python
data = pd.concat([data, features_data], axis=1, join='inner')
data
```

```python
data.to_csv('data/cleaned_data.csv')
```
